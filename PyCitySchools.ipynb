```python
#Dependencies & setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data,on=["school_name"])
school_data_complete.head()
```

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete["student_name"].count()
student_count
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = round(school_data_complete["math_score"].mean(),0)
average_math_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / student_count * 100
passing_math_percentage
```

```python
 # Calculate the percentage of students who passed reading  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / student_count * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total Schools": [school_count],
                                "Total Students": [student_count],
                               "Total Budget": [total_budget],
                               "Average Math Score":[average_math_score],
                               "% of Students Who Passed Math": [passing_math_percentage],
                               "% of Students Who Passed Reading": [passing_reading_percentage],
                               "Overall Passing Rate %": [overall_passing_rate]},index = ["Summary"])

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,.2f}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["% of Students Who Passed Math"] = district_summary["% of Students Who Passed Math"].map("{:,.2f}".format)
district_summary["% of Students Who Passed Reading"] = district_summary["% of Students Who Passed Reading"].map("{:,.2f}".format)
district_summary["Overall Passing Rate %"] = district_summary["Overall Passing Rate %"].map("{:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
# Use the code provided to select all of the school types
school_types = pd.DataFrame(school_data.set_index(["school_name"])["type"]).rename(columns={"type":"District Type"})
school_types
```

```python
# Calculate the total student count per school
per_school_counts = pd.DataFrame(school_data_complete.groupby(["school_name"]).count()["student_name"])
per_school_counts = per_school_counts.rename(columns={"school_name":"Name","student_name":"Population"})
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = pd.DataFrame(school_data[["school_name","budget"]])
per_school_capita = pd.merge(per_school_counts,per_school_budget,on=["school_name"])
per_school_capita["Per Capita"] = per_school_capita["budget"]/per_school_capita["Population"]
per_school_capita
```

```python
# Calculate the average test scores per school
per_school_math = pd.DataFrame(school_data_complete.groupby(["school_name"]).mean()["math_score"]).rename(columns={"math_score":"Avg Math Score"})
per_school_reading = pd.DataFrame(school_data_complete.groupby(["school_name"]).mean()["reading_score"]).rename(columns={"reading_score":"Avg Reading Score"})
```

```python
# Calculate the number of students per school with math scores of 70 or higher
school_students_passing_math =pd.DataFrame(school_data_complete[(school_data_complete["math_score"] >= 70)].groupby(["school_name"]).count()["math_score"]).rename(columns={"math_score":"Students Passing Math"})
school_students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
school_students_passing_reading = pd.DataFrame(school_data_complete[(school_data_complete["reading_score"]>=70)].groupby(["school_name"]).count()["reading_score"]).rename(columns={"reading_score":"Students Passing Reading"})
school_students_passing_reading 
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = pd.DataFrame(school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"] >= 70)].groupby(["school_name"]).count()["student_name"]).rename(columns={"student_name":"Students Passing Math & Reading"})
students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math_reading_rates =pd.merge(pd.merge(pd.merge(school_students_passing_math,school_students_passing_reading,on=["school_name"]),students_passing_math_and_reading,on=["school_name"]),per_school_counts,on=["school_name"])
per_school_passing_math_reading_rates["Student's Passing Math %"] = per_school_passing_math_reading_rates["Students Passing Math"] / per_school_passing_math_reading_rates["Population"] * 100
per_school_passing_math_reading_rates["Students Passing Reading %"] = per_school_passing_math_reading_rates["Students Passing Reading"] / per_school_passing_math_reading_rates["Population"] * 100
per_school_passing_math_reading_rates["Students Passing Math & Reading %"] = per_school_passing_math_reading_rates["Students Passing Math & Reading"] / per_school_passing_math_reading_rates["Population"] * 100
per_school_passing_math_reading_rates = per_school_passing_math_reading_rates.drop("Population",axis='columns')
per_school_passing_math_reading_rates
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary =pd.merge(pd.merge(pd.merge(pd.merge(school_types,per_school_capita,on=["school_name"]),per_school_math,on=["school_name"]),per_school_reading,on=["school_name"]),per_school_passing_math_reading_rates,on=["school_name"])
per_school_summary["budget"] = per_school_summary["budget"].map("${:,.2f}".format)
per_school_summary["Per Capita"] = per_school_summary["Per Capita"].map("${:,.0f}".format)
per_school_summary
```

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.
top_schools = per_school_summary.sort_values("Students Passing Math & Reading %", ascending = False)
top_schools.head(5)
```

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.
bottom_schools = per_school_summary.sort_values("Students Passing Math & Reading %")
bottom_schools.head(5)
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Avg Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Avg Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Student's Passing Math %"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Students Passing Reading %"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Students Passing Math & Reading"].mean()
```

```python
# Assemble into DataFrame
spending_summary = scores_by_spend = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    '% Passing Math': spending_passing_math,
    '% Passing Reading': spending_passing_reading,
    "Overall Passing Rate": overall_passing_spending})

# Display results
spending_summary
```
